```python
density_db = {'NGF': ('A1','nonVIP'),
              'PV': ('A1','PV'),
              'SOM': ('A1','SOM'),
              'VIP': ('A1','VIP')}
```

```python
densitites = {'NGF': 0,
              'PV': 0,
              'SOM': 0,
              'VIP': 0}

for Ipop, dens in density_db.items():
    for i in range(6):

        Icell = f'{Ipop}{i+1}'
        density_num = density[density_db[Ipop]][i]
        if i == 0:
            if 'NGF' not in Ipop:
                density_num = 0
        if i == 1:
            density_num *= 2
        densitites[Ipop] += density_num

        print(f'{Icell}, {density_num}')
        
        
```

```python
densitites['total'] = 0
for dens in densitites.values():
    densitites['total'] += dens
```
