```python
import matplotlib
import os
import json
import model_helpers as mh
from neuron import h, load_mechanisms
%matplotlib inline
```

### Define model

```python
# nmldb_id = "NMLCL000073"
model_source = 'NetPyNE'
model_name = 'E21'  # "L5PCbiophys"
model_version = 'NEURON'  # '1', '2', '3', '4' for Hay model otherwise ''

cell_model = 'HH3D'
cell_type = 'PYR'

hoc_name = 'geom'
```

### Define directories

```python
cwd = os.getcwd()
models_dir = os.path.join(cwd, 'models')
model_dir = os.path.join(models_dir, model_source, model_name)  # 'L5bPCmodelsEH')
hocs_dir = model_dir if 'biophys' not in model_name else os.path.join(model_dir,'models')
mod_dir = model_dir if 'biophys' not in model_name else os.path.join(model_dir, 'mod')
```

### Compile Mechanisms

```python
load_mechanisms(model_dir)
mh.compile_mechs(cwd, hocs_dir, mod_dir)
```

### Import cell into NetPyNE

```python
hoc_file = os.path.join(hocs_dir, hoc_name+'.hoc')
# morph_file = os.path.join(model_dir,'morphologies','cell1.asc')
```

```python
from netpyne import specs, sim

# Network parameters
netParams = specs.NetParams()
```

E21 from NetPyNE Tutorial 2021

```python
cell_label = model_name+'_hoc'

netParams.importCellParams(
    label=cell_label,
    conds={'cellType': cell_type, 'cellModel': cell_model},
    fileName=hoc_file,
    cellName = model_name,
    importSynMechs=False
)
```

```python
netParams.cellParams.keys()
```

```python
netParams.cellParams[cell_label]['secs']['soma']['geom']['L']
```

### Simulation Configuration

```python
## cfg
cfg = specs.SimConfig()					            # object of class SimConfig to store simulation configuration
cfg.duration = 2*1e3 						            # Duration of the simulation, in ms
cfg.dt = 0.01								                # Internal integration timestep to use
cfg.verbose = 1							                # Show detailed messages
cfg.recordTraces = {'V_soma':{'sec':'soma','loc':0.5,'var':'v'}}  # Dict with traces to record
cfg.recordStep = 0.01
cfg.filename = model_dir+'model_output_'+cell_label  			# Set file output name
cfg.saveJson = False
cfg.analysis['plotTraces'] = {'include': [0], 'saveFig': True} # Plot recorded traces for this list of cells
cfg.hParams['celsius'] = 36
```

```python
pop_label = cell_label+'_pop'
netParams.popParams[pop_label] = {'cellType': cell_type, 'numCells': 1, 'cellModel': cell_model}
```

```python
if 'NeuroML' in model_version:
    sim.importNeuroML2SimulateAnalyze(nml_file_name, cfg)
elif 'NEURON' in model_version:
    sim.createSimulateAnalyze(netParams = netParams, simConfig = cfg)
else:
    print('Invalid model version!')
```

### Plot Morphology

```python
sim.analysis.plotShape(includePre = [], includePost=[pop_label], showSyns=False, figSize=(4,9), dist=0.8, saveFig=True)
```
